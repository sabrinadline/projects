```python
#fungsi perhitungan peluang
def peluang (jumlah, total):
    hitung1 = "%.3f" % (jumlah/total)
    return hitung1

#fungsi perhitungan uji hipotesa selisih dua proporsi
def Zuji (p1, p2, n):
    hitung2 = "%.3f" % ((p1-p2)/(np.sqrt((p1*(1-p1)+p2*(1-p2))/n)))
    return hitung2

def Zuji2 (p1, p2, n1, n2):
    hitung3 = "%.3f" % ((p1-p2)/(np.sqrt(((p1*(1-p1))/n1)+((p2*(1-p2))/n2))))
    return hitung3

#fungsi perhitungan Confidence Interval
def CI (p, n, z):
    pbawah = "%.3f" % (p-(z*(np.sqrt((p*(1-p))/n))))
    patas = "%.3f" % (p+(z*(np.sqrt((p*(1-p))/n))))
    print (pbawah, " < p < ", patas)
```
